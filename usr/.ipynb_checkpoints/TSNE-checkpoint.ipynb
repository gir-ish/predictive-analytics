```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, roc_curve
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.metrics import roc_curve 
import sklearn.metrics  
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

# Load and preprocess the data
def load_and_preprocess_data(train_file, dev_file, test_file):
    train = pd.read_csv(train_file)
    dev = pd.read_csv(dev_file)
    test = pd.read_csv(test_file)
    
    X_train = train.iloc[:, 1:-1].values
    y_train = train['class'].values

    X_dev = dev.iloc[:, 1:-1].values
    y_dev = dev['class'].values

    X_test = test.iloc[:, 1:-1].values
    y_test = test['class'].values

    encoder = LabelEncoder()
    y_train = encoder.fit_transform(y_train)
    y_dev = encoder.transform(y_dev)
    y_test = encoder.transform(y_test)

    X_train = np.expand_dims(X_train, axis=2)
    X_dev = np.expand_dims(X_dev, axis=2)
    X_test = np.expand_dims(X_test, axis=2)
    
    return X_train, y_train, X_dev, y_dev, X_test, y_test

# Compute Equal Error Rate (EER)
def compute_eer(label, pred, positive_label=1):
    fpr, tpr, threshold = sklearn.metrics.roc_curve(label, pred, pos_label=positive_label)
    fnr = 1 - tpr
    eer_threshold = threshold[np.nanargmin(np.absolute((fnr - fpr)))]
    eer_1 = fpr[np.nanargmin(np.absolute((fnr - fpr)))]
    eer_2 = fnr[np.nanargmin(np.absolute((fnr - fpr)))]
    eer = (eer_1 + eer_2) / 2
    return eer

# Build the CNN model
def build_cnn_model(input_shape):
    model = Sequential([
        Conv1D(64, kernel_size=3, activation='relu', input_shape=input_shape),
        BatchNormalization(), 
        MaxPooling1D(pool_size=2),
        Dropout(0.2),   
        Conv1D(128, kernel_size=3, activation='relu'),
        BatchNormalization(),  
        MaxPooling1D(pool_size=2),
        Dropout(0.2),
        Flatten(),
        Dense(128, activation='relu'),
        BatchNormalization(),  
        Dropout(0.5),
        Dense(2, activation='softmax') 
    ])
    
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    return model

# Coreset selection using clustering and boundary point selection
def find_coreset(X, y, n_clusters=10):
    kmeans = KMeans(n_clusters=n_clusters)
    clusters = kmeans.fit_predict(X)

    coreset_data = []
    coreset_labels = []
    
    for i in range(n_clusters):
        cluster_data = X[clusters == i]
        cluster_labels = y[clusters == i]
        
        # If both classes are present in the cluster, find the closest points
        if len(np.unique(cluster_labels)) == 2:
            class_0_points = cluster_data[cluster_labels == 0]
            class_1_points = cluster_data[cluster_labels == 1]
            
            # Compute the distance between points of class 0 and class 1
            distances = cdist(class_0_points, class_1_points, metric='euclidean')
            
            # Find the minimum distance and corresponding points
            min_dist_idx = np.unravel_index(np.argmin(distances), distances.shape)
            closest_class_0_point = class_0_points[min_dist_idx[0]]
            closest_class_1_point = class_1_points[min_dist_idx[1]]
            
            # Add these points to the coreset
            coreset_data.append(closest_class_0_point)
            coreset_data.append(closest_class_1_point)
            coreset_labels.append(0)
            coreset_labels.append(1)
    
    return np.array(coreset_data), np.array(coreset_labels)

# Train and evaluate the CNN using coreset for both training and dev data
def train_and_evaluate_cnn_with_coreset(train_file, dev_file, test_file, epochs=10, batch_size=32, n_clusters=10):
    # Load and preprocess data
    X_train, y_train, X_dev, y_dev, X_test, y_test = load_and_preprocess_data(train_file, dev_file, test_file)
    
    # Find the coreset for both training and dev sets using clustering and boundary point selection
    coreset_X_train, coreset_y_train = find_coreset(X_train[:, :, 0], y_train, n_clusters)
    coreset_X_dev, coreset_y_dev = find_coreset(X_dev[:, :, 0], y_dev, n_clusters)
    
    # Reshape the coreset to match the CNN input shape
    coreset_X_train = np.expand_dims(coreset_X_train, axis=2)
    coreset_X_dev = np.expand_dims(coreset_X_dev, axis=2)

    # Build the CNN model
    model = build_cnn_model(input_shape=(coreset_X_train.shape[1], 1))
  
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
    
    # Train the model on the coreset from both training and dev data
    model.fit(coreset_X_train, coreset_y_train, validation_data=(coreset_X_dev, coreset_y_dev), epochs=epochs, batch_size=batch_size,
              callbacks=[reduce_lr])
    
    # Evaluate the model on the test set
    y_pred_prob = model.predict(X_test)[:, 1]  
    y_pred = np.argmax(model.predict(X_test), axis=1)
    
    # Print classification report
    print(classification_report(y_test, y_pred, target_names=['bonafide', 'spoof'], digits=4))

    # Compute Equal Error Rate (EER)
    eer = compute_eer(y_test, y_pred_prob, positive_label=1)
    eer_percentage = eer * 100  
    print(f'Equal Error Rate (EER): {eer_percentage:.4f}%')
    
    return model

# Example usage:
train_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_train_Wav2Vec2.csv'
dev_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_dev_Wav2Vec2.csv'
test_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_test_Wav2Vec2.csv'

# Call the modified function to train and evaluate the CNN model with a coreset on both training and dev data
cnn_model = train_and_evaluate_cnn_with_coreset(train_file, dev_file, test_file, epochs=100, batch_size=32, n_clusters=2)
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

# Function to select clusters and create the coreset based on their importance
def dynamic_coreset_selection(X, y, n_clusters=10, initial_clusters=5):
    kmeans = KMeans(n_clusters=n_clusters)
    clusters = kmeans.fit_predict(X)

    # Start by selecting a small set of clusters
    selected_clusters = np.random.choice(range(n_clusters), initial_clusters, replace=False)
    coreset_data = []
    coreset_labels = []

    for i in selected_clusters:
        cluster_data = X[clusters == i]
        cluster_labels = y[clusters == i]
        
        # If both classes are present, find the closest points
        if len(np.unique(cluster_labels)) == 2:
            class_0_points = cluster_data[cluster_labels == 0]
            class_1_points = cluster_data[cluster_labels == 1]
            
            # Compute the distance between points of class 0 and class 1
            distances = cdist(class_0_points, class_1_points, metric='euclidean')
            
            # Find the minimum distance and corresponding points
            min_dist_idx = np.unravel_index(np.argmin(distances), distances.shape)
            closest_class_0_point = class_0_points[min_dist_idx[0]]
            closest_class_1_point = class_1_points[min_dist_idx[1]]
            
            # Add these points to the coreset
            coreset_data.append(closest_class_0_point)
            coreset_data.append(closest_class_1_point)
            coreset_labels.append(0)
            coreset_labels.append(1)

    return np.array(coreset_data), np.array(coreset_labels), clusters, kmeans

# Updated training function with dynamic coreset selection
def train_and_evaluate_cnn_with_dynamic_coreset(train_file, dev_file, test_file, epochs=10, batch_size=32, n_clusters=10, initial_clusters=5):
    # Load and preprocess data
    X_train, y_train, X_dev, y_dev, X_test, y_test = load_and_preprocess_data(train_file, dev_file, test_file)
    
    # Initial coreset selection using clustering
    coreset_X_train, coreset_y_train, clusters, kmeans = dynamic_coreset_selection(X_train[:, :, 0], y_train, n_clusters, initial_clusters)
    
    # Reshape the coreset to match CNN input shape
    coreset_X_train = np.expand_dims(coreset_X_train, axis=2)

    # Build the CNN model
    model = build_cnn_model(input_shape=(coreset_X_train.shape[1], 1))

    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    
    # Train model on the initial coreset and monitor validation accuracy
    history = model.fit(coreset_X_train, coreset_y_train, validation_data=(X_dev, y_dev), epochs=epochs, batch_size=batch_size, callbacks=[reduce_lr, early_stop])
    
    # Track validation loss and accuracy
    best_val_accuracy = max(history.history['val_accuracy'])
    
    # Dynamically add more clusters if validation performance is not improving
    remaining_clusters = list(set(range(n_clusters)) - set(clusters))
    for cluster_id in remaining_clusters:
        # Extract points from the new cluster and add to the training set
        cluster_data = X_train[clusters == cluster_id, :, 0]
        cluster_labels = y_train[clusters == cluster_id]
        
        # Add the new cluster points to the coreset
        coreset_X_train = np.concatenate([coreset_X_train, np.expand_dims(cluster_data, axis=2)])
        coreset_y_train = np.concatenate([coreset_y_train, cluster_labels])
        
        # Train the model again with the new coreset
        history = model.fit(coreset_X_train, coreset_y_train, validation_data=(X_dev, y_dev), epochs=epochs, batch_size=batch_size, callbacks=[reduce_lr, early_stop])
        
        # Update validation accuracy and check if further improvement is happening
        new_val_accuracy = max(history.history['val_accuracy'])
        if new_val_accuracy <= best_val_accuracy:
            print("No further improvement in validation accuracy, stopping.")
            break
        best_val_accuracy = new_val_accuracy

    # Evaluate the model on the test set
    y_pred_prob = model.predict(X_test)[:, 1]  
    y_pred = np.argmax(model.predict(X_test), axis=1)
    
    # Print classification report
    print(classification_report(y_test, y_pred, target_names=['bonafide', 'spoof'], digits=4))

    # Compute Equal Error Rate (EER)
    eer = compute_eer(y_test, y_pred_prob, positive_label=1)
    eer_percentage = eer * 100  
    print(f'Equal Error Rate (EER): {eer_percentage:.4f}%')
    
    return model

# Example usage:
train_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_train_Wav2Vec2.csv'
dev_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_dev_Wav2Vec2.csv'
test_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_test_Wav2Vec2.csv'

# Call the updated function to train and evaluate the CNN model with dynamic coreset selection
cnn_model = train_and_evaluate_cnn_with_dynamic_coreset(train_file, dev_file, test_file, epochs=100, batch_size=32, n_clusters=5, initial_clusters=2)
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np

# Function to select more points from each cluster (e.g., top N closest pairs)
def find_coreset_with_multiple_points(X, y, n_clusters=100, points_per_cluster=5):
    kmeans = KMeans(n_clusters=n_clusters)
    clusters = kmeans.fit_predict(X)

    coreset_data = []
    coreset_labels = []

    for i in range(n_clusters):
        cluster_data = X[clusters == i]
        cluster_labels = y[clusters == i]
        
        # If both classes are present in the cluster, find the closest points
        if len(np.unique(cluster_labels)) == 2:
            class_0_points = cluster_data[cluster_labels == 0]
            class_1_points = cluster_data[cluster_labels == 1]
            
            # Compute the distance between points of class 0 and class 1
            distances = cdist(class_0_points, class_1_points, metric='euclidean')
            
            # Find the top N closest pairs of points
            closest_pairs_indices = np.unravel_index(np.argsort(distances, axis=None)[:points_per_cluster], distances.shape)
            
            # Add these points to the coreset
            for idx in range(points_per_cluster):
                closest_class_0_point = class_0_points[closest_pairs_indices[0][idx]]
                closest_class_1_point = class_1_points[closest_pairs_indices[1][idx]]
                coreset_data.append(closest_class_0_point)
                coreset_data.append(closest_class_1_point)
                coreset_labels.append(0)
                coreset_labels.append(1)
    
    return np.array(coreset_data), np.array(coreset_labels)

# Updated training function with dynamic coreset selection
def train_and_evaluate_cnn_with_large_coreset(train_file, dev_file, test_file, epochs=10, batch_size=32, n_clusters=100, points_per_cluster=5):
    # Load and preprocess data
    X_train, y_train, X_dev, y_dev, X_test, y_test = load_and_preprocess_data(train_file, dev_file, test_file)
    
    # Find the coreset using clustering and select multiple points per cluster
    coreset_X_train, coreset_y_train = find_coreset_with_multiple_points(X_train[:, :, 0], y_train, n_clusters, points_per_cluster)
    
    # Reshape the coreset to match CNN input shape
    coreset_X_train = np.expand_dims(coreset_X_train, axis=2)

    # Build the CNN model
    model = build_cnn_model(input_shape=(coreset_X_train.shape[1], 1))

    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
    #early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    
    # Train the model on the coreset and monitor validation accuracy
    history = model.fit(coreset_X_train, coreset_y_train, validation_data=(X_dev, y_dev), epochs=epochs, batch_size=batch_size, callbacks=[reduce_lr])
    
    # Evaluate the model on the test set
    y_pred_prob = model.predict(X_test)[:, 1]  
    y_pred = np.argmax(model.predict(X_test), axis=1)
    
    # Print classification report
    print(classification_report(y_test, y_pred, target_names=['bonafide', 'spoof'], digits=4))

    # Compute Equal Error Rate (EER)
    eer = compute_eer(y_test, y_pred_prob, positive_label=1)
    eer_percentage = eer * 100  
    print(f'Equal Error Rate (EER): {eer_percentage:.4f}%')
    
    return model

# Example usage:
train_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_train_Wav2Vec2.csv'
dev_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_dev_Wav2Vec2.csv'
test_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_test_Wav2Vec2.csv'

# Call the function to train and evaluate the CNN model with a larger coreset (1000 data points)
cnn_model = train_and_evaluate_cnn_with_large_coreset(train_file, dev_file, test_file, epochs=100, batch_size=32, n_clusters=100, points_per_cluster=5)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, roc_curve
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.metrics import roc_curve, silhouette_score
import sklearn.metrics  
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

# Load and preprocess the data
def load_and_preprocess_data(train_file, dev_file, test_file):
    train = pd.read_csv(train_file)
    dev = pd.read_csv(dev_file)
    test = pd.read_csv(test_file)
    
    X_train = train.iloc[:, 1:-1].values
    y_train = train['class'].values

    X_dev = dev.iloc[:, 1:-1].values
    y_dev = dev['class'].values

    X_test = test.iloc[:, 1:-1].values
    y_test = test['class'].values

    encoder = LabelEncoder()
    y_train = encoder.fit_transform(y_train)
    y_dev = encoder.transform(y_dev)
    y_test = encoder.transform(y_test)

    X_train = np.expand_dims(X_train, axis=2)
    X_dev = np.expand_dims(X_dev, axis=2)
    X_test = np.expand_dims(X_test, axis=2)
    
    return X_train, y_train, X_dev, y_dev, X_test, y_test

# Compute Equal Error Rate (EER)
def compute_eer(label, pred, positive_label=1):
    fpr, tpr, threshold = sklearn.metrics.roc_curve(label, pred, pos_label=positive_label)
    fnr = 1 - tpr
    eer_threshold = threshold[np.nanargmin(np.absolute((fnr - fpr)))]
    eer_1 = fpr[np.nanargmin(np.absolute((fnr - fpr)))]
    eer_2 = fnr[np.nanargmin(np.absolute((fnr - fpr)))]
    eer = (eer_1 + eer_2) / 2
    return eer

# Build the CNN model
def build_cnn_model(input_shape):
    model = Sequential([
        Conv1D(64, kernel_size=3, activation='relu', input_shape=input_shape),
        BatchNormalization(), 
        MaxPooling1D(pool_size=2),
        Dropout(0.2),   
        Conv1D(128, kernel_size=3, activation='relu'),
        BatchNormalization(),  
        MaxPooling1D(pool_size=2),
        Dropout(0.2),
        Flatten(),
        Dense(128, activation='relu'),
        BatchNormalization(),  
        Dropout(0.5),
        Dense(2, activation='softmax') 
    ])
    
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    return model

# Function to automatically determine the best number of clusters using Silhouette score
def determine_optimal_clusters(X, max_clusters=20):
    best_n_clusters = 2
    best_silhouette_score = -1
    for n_clusters in range(2, max_clusters):
        kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(X)
        silhouette_avg = silhouette_score(X, kmeans.labels_)
        if silhouette_avg > best_silhouette_score:
            best_silhouette_score = silhouette_avg
            best_n_clusters = n_clusters
    return best_n_clusters

# Function to select points based on an adaptive distance threshold
def find_coreset_with_adaptive_points(X, y, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters)
    clusters = kmeans.fit_predict(X)

    coreset_data = []
    coreset_labels = []

    for i in range(n_clusters):
        cluster_data = X[clusters == i]
        cluster_labels = y[clusters == i]

        # If both classes are present in the cluster
        if len(np.unique(cluster_labels)) == 2:
            class_0_points = cluster_data[cluster_labels == 0]
            class_1_points = cluster_data[cluster_labels == 1]

            # Compute distances between class 0 and class 1 points
            distances = cdist(class_0_points, class_1_points, metric='euclidean')
            avg_distance = np.mean(distances)
            
            # Select points adaptively based on distance (e.g., select points closer than avg distance)
            points_per_cluster = np.sum(distances < avg_distance)
            
            # Select those closest points
            closest_pairs_indices = np.unravel_index(np.argsort(distances, axis=None)[:points_per_cluster], distances.shape)

            for idx in range(points_per_cluster):
                closest_class_0_point = class_0_points[closest_pairs_indices[0][idx]]
                closest_class_1_point = class_1_points[closest_pairs_indices[1][idx]]
                coreset_data.append(closest_class_0_point)
                coreset_data.append(closest_class_1_point)
                coreset_labels.append(0)
                coreset_labels.append(1)

    return np.array(coreset_data), np.array(coreset_labels)

# Train and evaluate the CNN using the adaptive coreset
def train_and_evaluate_cnn_with_adaptive_coreset(train_file, dev_file, test_file, epochs=10, batch_size=32, max_clusters=20):
    # Load and preprocess data
    X_train, y_train, X_dev, y_dev, X_test, y_test = load_and_preprocess_data(train_file, dev_file, test_file)
    
    # Automatically determine the optimal number of clusters
    optimal_clusters = determine_optimal_clusters(X_train[:, :, 0], max_clusters=max_clusters)
    print(f"Optimal number of clusters determined: {optimal_clusters}")

    # Find the coreset adaptively based on the selected clusters and distance thresholds
    coreset_X_train, coreset_y_train = find_coreset_with_adaptive_points(X_train[:, :, 0], y_train, optimal_clusters)
    
    # Reshape the coreset to match CNN input shape
    coreset_X_train = np.expand_dims(coreset_X_train, axis=2)

    # Build and train the CNN model
    model = build_cnn_model(input_shape=(coreset_X_train.shape[1], 1))

    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
    history = model.fit(coreset_X_train, coreset_y_train, validation_data=(X_dev, y_dev), epochs=epochs, batch_size=batch_size, callbacks=[reduce_lr])
    
    # Evaluate the model on the test set
    y_pred_prob = model.predict(X_test)[:, 1]  # Get the probability of the positive class
    y_pred = np.argmax(model.predict(X_test), axis=1)  # Get the predicted class
    print(classification_report(y_test, y_pred, target_names=['bonafide', 'spoof'], digits=4))

    eer = compute_eer(y_test, y_pred_prob, positive_label=1)
    eer_percentage = eer * 100  
    print(f'Equal Error Rate (EER): {eer_percentage:.4f}%')
    
    return model

# Example usage:
train_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_train_Wav2Vec2.csv'
dev_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_dev_Wav2Vec2.csv'
test_file = '/home/girish/Girish/RESEARCH/EMO-FAKE/FEATURES/Wav2Vec2/English_test_Wav2Vec2.csv'

# Call the function to train and evaluate the CNN model with adaptive coreset selection
cnn_model = train_and_evaluate_cnn_with_adaptive_coreset(train_file, dev_file, test_file, epochs=100, batch_size=32, max_clusters=20)
```
